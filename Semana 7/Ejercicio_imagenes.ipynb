```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('los-10-sonidos-principales-del-perro.jpg')

# Get the image dimensions
height, width = image.shape[:2]

# Calculate the center of the image
center = (width // 2, height // 2)

# Define the rotation matrix
angle = 45
M = cv2.getRotationMatrix2D(center, angle, 1.0)

# Apply the rotation to the image
rotated = cv2.warpAffine(image, M, (width, height))

# Show the rotated image
cv2.imshow('Rotated', rotated)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('los-10-sonidos-principales-del-perro.jpg')

# Get the image dimensions
height, width = image.shape[:2]

# Define the translation parameters
tx = 100
ty = 50

# Create the translation matrix
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image
translated = cv2.warpAffine(image, M, (width, height))

# Show the translated image
cv2.imshow('Translated', translated)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python

import cv2
import numpy as np
# Cargar la imagen
image = cv2.imread('los-10-sonidos-principales-del-perro.jpg')
# Definir el kernel para el filtro de afilado
kernel = np.array([[-1, -1, -1],
                    [-1, 9, -1],
                    [-1, -1, -1]])
# Aplicar el filtro de afilado para realzar los detalles 
sharpened = cv2.filter2D(image, -1, kernel)
# Mostrar la imagen realzada
cv2.imshow('Realce', sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

import cv2
# Cargar la imagen en escala de grises
image = cv2.imread('los-10-sonidos-principales-del-perro.jpg', cv2.IMREAD_GRAYSCALE)
# Aplicar el operador Sobel para detectar bordes

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5) 
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
# Combinar las respuestas en magnitud
edges = cv2.magnitude (sobelx, sobely)
# Normalizar los valores para mostrar la imagen correctamente 
edges = cv2.normalize (edges, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
# Mostrar la imagen con bordes detectados
cv2.imshow('Detección de Bordes', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

import numpy as np

faceClassif = cv2.CascadeClassifier('haarcascade_frontalface.xml')


# Cargar la imagen
image = cv2.imread('perros.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Detección de rostros de gatos:
faces = faceClassif.detectMultiScale(gray,
scaleFactor=1.1,
minNeighbors=5,
minSize=(30,30),
maxSize=(200,200))
print(faces)

# Dibujar rectángulos alrededor de los rostros detectados 
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y), (x+w, y+h), (0,255,0),2)


# Mostrar la imagen con los contornos resaltados 
cv2.imshow('Objects Detected', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
